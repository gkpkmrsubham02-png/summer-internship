```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\dell\\Downloads\\archive\\zomato.csv")
df.head(5)
```

```python
df.drop(["url","reviews_list","menu_item","listed_in(city)"],axis=1,inplace=True)
df.head()
```

```python
df.columns
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["rate"].unique()
```

```python
#removing "New","-"and "/5"from rate column
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
      value=str(value).split('/')
      value=value[0]
      return float(value)
df["rate"]=df["rate"].apply(handlerate)
df["rate"].head()
```

```python
df["rate"].unique()
```

```python
df["rate"].fillna(df["rate"].mean(),inplace=True)
df["rate"].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head(2)
```
